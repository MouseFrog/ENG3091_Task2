```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data and force it to recognize commas
# 'skipinitialspace' helps if there are spaces after your commas in C++
df = pd.read_csv('prices.csv', sep=',', skipinitialspace=True)

# 2. Clean up column names (removes any hidden newlines or spaces)
df.columns = df.columns.str.strip()

# 3. Create the single plot
plt.figure(figsize=(5, 3))
plt.scatter(df['Number of Bedrooms'], df['Single Variable Price'], alpha=0.6, color='royalblue', edgecolors='black')
# 4. Labels and Title
plt.title('Single Prices vs. Number of Bedrooms', fontsize=14)
plt.xlabel('Number of Bedrooms', fontsize=12)
plt.ylabel('Price (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.figure(figsize=(5, 3))
plt.scatter(df['Land Size'], df['Multi Variable Price'], alpha=0.6, color='royalblue', edgecolors='black')
# 4. Labels and Title
plt.title('Multi Price vs. Number of Bedrooms', fontsize=14)
plt.xlabel('Number of Bedrooms ', fontsize=12)
plt.ylabel('Price (£)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.figure(figsize=(5, 3))
plt.scatter(df['Number of Bedrooms'], df['Land Size'], alpha=0.6, color='royalblue', edgecolors='black')
# 4. Labels and Title
plt.title('Land Size vs. Number of Bedrooms', fontsize=14)
plt.xlabel('Number of Bedrooms', fontsize=12)
plt.ylabel('Land Size', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
# 5. Show the plot
plt.show()
```
